## **I. Linear Algebra là gì?**

### 1. Khái niệm

Đại số tuyến tính là khái niệm số học bao gồm các thực tế "không phải số" như `x`. Thể hiện qua phương trình: 

$2x + 5 = 25$ 

Có thể thực hiện biến đổi để ra kết quả của $x$ như sau:

<=> $2x + 5 - 5 = 25 - 5$

<=> $2x = 20$

<=> $2x / 2 = 20 / 2$

<=> $x = 10$

Trong trường hợp exponential term (số mũ), không còn gọi là Linear Algebra nữa, ví dụ: 

$2x^2 + 5 = 25$

$2 \sqrt x + 5 = 25$

Hay nói cách khác, Linear Algebra là việc đi giải hệ phương trình tuyến tính để tìm các ẩn số.

### 2. Ví dụ 

- Cảnh sát trưởng có 1 chiếc xe chạy tốc độ max = 180 km/h

- Tên cướp có chiếc xe chạy tốc độ max = 150 km/h, và hắn đã chạy xe trước cảnh sát trưởng 5 phút 

- Sau bao lâu thì cảnh sát trưởng bắt được tên cướp ? và họ đã đi được km tính tới thời điểm đó?

> Có thể mô tả thông qua hình ảnh sau:

> Để giải, đặt phương trình như sau: 

Gọi 

s = quãng đường cảnh sát trưởng và tên cướp chạy tới khi bị bắt

Công thức $v = {s \over t}$ <=> $t = {s\over v}$

Thời gian cướp chạy: $t1 = {s \over 150}$

Thời gian cảnh sát chạy: $t2 = {s \over 180}$

Trong đó, t1 - t2 = 5 phút ~ 5/60 giờ

<=> ${s \over 150} - {s \over 180} = {5 \over 60} $

<=> $s = 75 (km)$

<=> Thời gian để cảnh sát bắt được tên cướp là ${75 \over 180} \ giờ = 25 \ phút$ kể từ lúc cảnh sát xuất phát

> Lưu ý: 

- Nếu cảnh sát và cướp có cùng tốc độ (same slope), thì không bao giờ bắt được cướp ==> 'No solution' hay 'Phương trình vô nghiệm'

- Nếu cảnh sát có cùng tốc độ và xuất phát cùng thời gian (same slope + intercept), thì lúc nào cũng bắt được cướp ==> 'Infinite solution' hay 'Phương trình vô số nghiệm'

### 3. Tổng quát

Từ đó, có thể thấy, với Linear Algebra, chỉ có 3 kết quả:

- 1 nghiệm

- Vô nghiệm

- Vô số nghiệm

Trong hệ phương trình đã cho, có thể:

- Có rất nhiều phương trình

- Có rất nhiều biến trong mỗi phương trình

Ví dụ, đối với phương trình giá trị của 1 căn nhà bất kỳ sẽ là như sau:

$$
\text{house\_price} = \text{intercept} + \text{distance\_to\_school} \cdot x_1 + \text{num\_bedrooms} \cdot x_2 + \ldots + \text{m\_features} \cdot x_m
$$

và đối với bất cứ căn nhà thứ i nào trong bộ dữ liệu, ta có hệ phương trình sau:

$$
\begin{bmatrix}
y_1 \\
y_2 \\
\vdots \\
y_n \\
\end{bmatrix}
=
\begin{bmatrix}
a + b x_{1,1} + c x_{1,2} + \cdots + m x_{1,m} \\
a + b x_{2,1} + c x_{2,2} + \cdots + m x_{2,m} \\
\vdots \\
a + b x_{n,1} + c x_{n,2} + \cdots + m x_{n,m} \\
\end{bmatrix}
$$

và $y_i$ là giá nhà của căn nhà i, các giá trị $x_{i,1}$ tới $x_{i,m}$ là các features của căn nhà đó
